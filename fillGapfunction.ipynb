```python
def gapfill(df):
  df=df[['Position X','Position Y','Position Z','Time','TrackID']]
  df=df.set_index('Time').reindex(range(df.Time.min(),df.Time.max()+1)).fillna(0.0).rename_axis('Time').reset_index() #fill the gap with zero for further condition statement
  start,end=0,0
  while start<len(df) and end<len(df):
    if any(df.loc[start][1:4]):
      start+=1
      end+=1
    elif not all(df.loc[end][1:4]):
      end+=1
    else:
      for i in range(start,end):
        df.iloc[i,1:4]=[df.iloc[start-1,1]+(df.iloc[end,1]-df.iloc[start-1,1])/(end-start+1)*(i-start+1),df.iloc[start-1,2]+(df.iloc[end,2]-df.iloc[start-1,2])/(end-start+1)*(i-start+1),df.iloc[start-1,3]+(df.iloc[end,3]-df.iloc[start-1,3])/(end-start+1)*(i-start+1)]
      start=end
  return df
```
