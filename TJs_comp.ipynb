```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_scalebar.scalebar import ScaleBar
```

Plot 2D M2 trajectories

```python
M2=[2,3,1,5,6,7,11,27]
sns.set_style('white')
plt.figure(figsize=(15,12),dpi=300)
for i in M2:
  df=pd.read_csv('/content/drive/MyDrive/2D_sperm_hyper_data/spermtrackingdata/M2_sperm_{}.csv'.format(i))
  plt.plot(df['Position X'],df['Position Y'],linewidth=2.5,alpha=0.7)
  plt.tick_params(axis='both',labelbottom=False,labelleft=False)
scalebar = ScaleBar(1,"um",location="lower left",scale_loc="top",pad=0.5,border_pad=1) # 1 pixel = 1 um
plt.gca().add_artist(scalebar)
plt.savefig('/content/drive/MyDrive/2D_sperm_hyper_data/M2_TJs')
plt.show()
```
